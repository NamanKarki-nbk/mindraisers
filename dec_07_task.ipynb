8. Create a class with methods to perform basic arithmetic operations and call them.

```python
class Calculator:
    def sum(self, a, b):
        return a + b
    
    def sub(self, a, b):
        return a - b
    
    def division(self, a, b):
        try:
            return a / b
        except ZeroDivisionError:
            return "Error: Division by zero"
    
    def multiply(self,a,b):
        return a * b
    
    def modulus(self, a, b):
        try:
            return a % b
        except ZeroDivisionError:
            return "Error: Modulus by zero."
    
    def power(self, a, b):
        return a ** b



calc = Calculator()
print(calc.sum(45, 5))       
print(calc.sub(100, 50))
print (calc.multiply(2,3))     
print(calc.division(50, 0))  
print(calc.modulus(50, 2))   
print(calc.power(2, 3))     
```


9. Create a class with a constructor that initializes two attributes and displays them.

```python
class Calculator:
    
    def __init__(self,a,b):
        self.a = a
        self.b = b
        print(f"First attribute: {self.a} \nSecond attribute: {self.b}")
        
    def sum(self):
        return self.a + self.b
    
    def sub(self):
        return self.a - self.b
    
    def multiply(self):
        return self.a * self.b
    
    def division(self):
        try:
            return self.a / self.b
        except ZeroDivisionError:
            return "Error: Division by zero"
    
    def modulus(self):
        try:
            return self.a % self.b
        except ZeroDivisionError:
            return "Error: Modulus by zero."
    
    def power(self):
        return self.a ** self.b

calc = Calculator(3, 2)
print("Sum:", calc.sum())          
print("Sub:", calc.sub())          
print("Multiply:", calc.multiply())
print("Division:", calc.division())
print("Modulus:", calc.modulus()) 
print("Power:", calc.power())

  
```


10. Create a parent class and a child class where the child overrides one method.

```python
class Parent():
    def __init__(self):
        print("This is The parent class")
        
    def view(self):
        print("I am view from parent class")
        

class Child(Parent):
    def __init__(self):
        super().__init__()
        print("This is the child class")
        
    def view(self):
        print("I am view form child class")
        

p = Parent()
p.view()

c = Child()
c.view()
```
