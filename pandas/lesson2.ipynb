```python
# print(df)
df.dropna(inplace=True)  # Drop rows with any null values
df.loc["Kiran","Column2"] = 94.0
df.loc["Kiran","Column3"] = "C"
print(df)
pivot_df = df.pivot(index='Column1', columns='Column3', values='Column2')
print(pivot_df)
```

```python
# Update Ram's Column2
df.loc['Ram', 'Column2'] = 100

# Recreate the pivot table after update
df_pivot = pd.pivot_table(
    data=df,
    index='Names',
    values='Column2',
    aggfunc='sum'
)

print(df_pivot.loc["Ram"])  # Should now show 100000
```

```python
print(df)
pivot_table_df = df.pivot_table(index='Column3', columns='Column5', values='Column2', aggfunc='mean')
print(pivot_table_df)
```

============================= DataFrame.dropna() function

this is a dropna() functions:
---------------DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)



axis: It determines the axis to remove. Set it to 1 or column to remove columns containing missing values. By default, it removes rows with NA from DataFrame.
how: It takes the following inputs:
any: This is the default case to drop the column if it has at least one value missing.
all: Drop the column only if it has all the values as NA.
thresh: It applies a condition to drop the columns only if it does not contain the required number of values. It takes an int as input.
subset: While dropping columns, it is used to specify the list of rows to be considered to find NA.
inplace: It is used to specify whether to return a new DataFrame or update an existing one. It is a boolean flag with default False.

![alt text](image-1.png)

```python
df.dropna(axis=0, how='any', subset=None, inplace=True)
df

```

| Parameter   | Description                                                                                                                                                          | Example / Notes                                                                                                                  |
| ----------- | -------------------------------------------------------------------------------------------------------------------------------------------------------------------- | -------------------------------------------------------------------------------------------------------------------------------- |
| **axis**    | Determines if you drop **rows** or **columns** with missing values. `0` = rows, `1` = columns                                                                        | `df.dropna(axis=0)`  drop rows with NaN<br>`df.dropna(axis=1)`  drop columns with NaN                                          |
| **how**     | Determines if a row/column is dropped **when it has any or all NaNs**. Options:<br>`any`  drop if **any NaN** exists<br>`all`  drop only if **all values are NaN** | `df.dropna(how='any')`  default, drops rows with any missing value<br>`df.dropna(how='all')`  only drops rows completely empty |
| **thresh**  | Minimum number of **non-NaN values required to keep** the row/column                                                                                                 | `df.dropna(thresh=2)`  keep rows with **at least 2 non-NaN values**                                                             |
| **subset**  | Only consider certain columns for checking NaN                                                                                                                       | `df.dropna(subset=['Age', 'Score'])`  only check these columns                                                                  |
| **inplace** | If `True`, modifies the original DataFrame. If `False` (default), returns a new DataFrame                                                                            | `df.dropna(inplace=True)`  original `df` is updated<br>`df_clean = df.dropna()`  creates new cleaned df                        |
