Data Loading and Sanity CHecks



```python
import pandas as pd
```

```python
hr_attendance_df = pd.read_csv('hr_attendance.csv', sep=',')
hr_employee_df = pd.read_csv('hr_employees.csv', sep=",")
hr_performance_df = pd.read_csv('hr_performance.csv', sep=",")
hr_salary_history_df = pd.read_csv('hr_salary_history.csv', sep=",")
```

```python
#removing duplicates row  missing values
hr_attendance_df = hr_attendance_df.drop_duplicates()
hr_employee_df = hr_employee_df.dropna(axis=0)
hr_performance_df = hr_performance_df.drop_duplicates()
```

```python
def sanity_check(df):
    invalid_values = ["NA", "N/A", "null", "None", "-"]
    print("\nShape of DataFrame:", df.shape)
    print("\n Columns and Data types: ", df.dtypes)
    print("\n Missing values in each column:", df.isnull().sum())
    print("\n Duplicate rows:", df.duplicated().sum())
    for col in df.columns:
        if df[col].isin (invalid_values).any():
            count = df[col].isin(invalid_values).sum()
            print(f"\n Invalid values found in column {col}")
            
    
```

```python
sanity_check(hr_attendance_df)
```

```python
sanity_check(hr_employee_df)
```

```python
sanity_check(hr_performance_df)
```

```python
sanity_check(hr_salary_history_df)
```

Indexing adn Time Alignment

```python
#setting emp_id as index for hr_employee_df
hr_employee_df = hr_employee_df.set_index("emp_id").sort_index()
```

```python
#settign multiindex on salary_history_df => emp_id and effective_date and attendance => emp_id and date
hr_salary_history_df = hr_salary_history_df.set_index(["emp_id", "effective_date"]).sort_index()
hr_attendance_df = hr_attendance_df.set_index(["emp_id", "date"]).sort_index()
```
