Task: Perform array slicing

```python
import numpy as np
```

```python
arr = np.arange(1,15)
print(arr)
```

```python
print(arr[0:16])
```

```python
print(arr[0:])
```

```python
print(arr[9:-1])
```

```python
print(arr[::])
```

```python
print(arr[:-1])
```

```python
print(arr[-1])
```

```python
print(arr[::-1])
```

```python
print(arr[12:])
```
