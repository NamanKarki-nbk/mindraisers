```python
import numpy as np
```

```python
#1. Create a 3x3 matrix with values from 1 to 9(2d array):

arr = np.arange(1,10).reshape(3,3)
print(arr)
```

```python
#2. Create an array of 10 random floats between 0 and 1, and round to 2 decimal places.

arr = np.random.rand(10).round(2)
print(arr)
```

```python
#3.Find all even numbers in the array [1, 3, 6, 9, 12, 15] using Boolean indexing

arr = np.array([1, 3, 6, 9, 12, 15])
mask = arr % 2 == 0
print(arr[mask])
```

```python
#4.Create a 4x4 matrix filled with 7
arr = np.full((4,4), fill_value=7)
print(arr)
```

Create a 3D array of shape (3, 4, 5) filled with random integers between 1 and 100.
Perform the following tasks:

```python
arr = np.random.randint(1,100, size=(3,4,5))
print(arr)
```

```python
#Replace all elements greater than the mean of the entire array with 0.

mean = arr.mean()
arr[arr > mean] = 0
print(mean)
print(arr)
```

```python
#Calculate the sum along axis=1 (i.e., sum across the 5 columns in each depth slice).
sum = arr.sum(axis =1)
print(sum)
```

```python
#Extract a new 2D array that contains only the first and last columns from each row of each depth slice.
arr2D = arr[:,:,[0,-1]]
print(arr2D)
```
