Task: Perform operations on 2d and 3d arrays

```python
#basic operations on 2d arrays

#uisng arange to create an array and reshaping it to 2D
arr = np.arange(20)
print(f"\nOriginal Array: {arr}")
print(f"\nReshaped 2D array: \n{arr.reshape(4,5)}")
```

```python
#using linspace to create an array and reshaping it to 2D
arr = np.linspace(0,19,20)
print(f"\nOriginal Array: {arr}")
print(f"\nReshaped Array: \n{arr.reshape(4,5)}")
```

```python
#using random.rand / random.randn
arr = np.random.rand(3,3)
print(f"\nArray:\n {arr}")
```

```python
#arithmentic operations for 2d array
arr1 = np.random.randn(2,2)
arr2 = np.random.randn(2,2)

#addition
print(f"Addition: {arr1+arr2}")
#sub
print(f"Sub: {arr1-arr2}")
#mul
print(f"Mul: {arr1*arr2}")
#division
print(f"Divison: {arr1/arr2}")
#mod
print(f"MOd: {arr1%arr2}")
#power
print(f"Power: {arr1 ** 2}")

```
