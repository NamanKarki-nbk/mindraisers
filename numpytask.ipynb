```python
import numpy as np
```

```python
# 1. Create a NumPy array of numbers from 10 to 50
arr1 = np.arange(10, 51)
print("Array :", arr1)
```

```python
# 2. Generate a 3×3 identity matrix
identity_matrix = np.eye(3,3)
identity_matrix
```

```python
# 3. Create a NumPy array of 10 random integers between 1 and 100
random_integers = np.random.randint(1, 101, size=10)
random_integers
```

```python
# 4. Find the maximum and minimum of an array
max_value = np.max(random_integers)
min_value = np.min(random_integers)
print("Maximum value:", max_value)
print("Minimum value:", min_value)
```

```python
# 5. Reshape a 1D array of 12 elements into a 3×4 matrix
random_integers = np.random.randint(1, 101, size=12)
reshaped_matrix = random_integers.reshape(3, 4)
reshaped_matrix
```

```python
# 6. Create an array of all even numbers between 50 and 100
even_numbers = np.arange(50, 101, 2)
even_numbers
```

```python
# 7. Compute the square of each element in [1,2,3,4,5]
array = np.array([1, 2, 3, 4, 5])
squared_array = np.square(array)
squared_array
```

```python
# 8. Create a 5×5 array of random numbers and find its mean
random_array = np.random.rand(5, 5)
mean_value = np.mean(random_array)
mean_value
```

```python
# 9. Find the index of the maximum element in [10, 20, 50, 40, 30]
array = np.array([10, 20, 50, 40, 30])
array.argmax()
```

```python
# 10. Create two arrays and perform element-wise operation
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])
sum_array = array1 + array2
product_array = array1 * array2
diff_array = array2 - array1
div_array = array2 / array1
mod_array = array2 % array1
sum_array, product_array, diff_array, div_array, mod_array
```
